# Check whether all the zones generate the same amount of sales.

```python
from scipy.stats import f_oneway
from scipy import stats
```

H0: Same amount of sales were generated by all zones
H1: Different amount of sales were generated by all zones

we are setting alpha to 5% so it is 0.05

```python
F, p = stats.f_oneway(df['Zone - A'], df['Zone - B'],df['Zone - C'],df['Zone - D'])
```

```python
alpha=.05
print("p value :", p)
if p < alpha:
    print("Reject H0")
else:
    print("Fail to reject H0.")
```

```python
f_oneway(df['Zone - A'], df['Zone - B'],df['Zone - C'],df['Zone - D'])
```

here since p value is less than alpha we reject null hypothesis and alternate hypothesis is failed to be rejected. so same amount of sales are not generated by all the zones.
